```python
# imports and packages
import pandas as pd
import numpy as np
```

```python
# loading dataset
df = pd.read_csv(r'raw\electricity-sample-dataset.csv')
```

```python
df.head()
```

```python
# storage of means
means_data = {col: [] for col in df.columns}
```

```python
np.random.seed(42)

# sampling means
for _ in range(50): # 50 times
    for c in df.columns:
        sample = np.random.choice(df[c], size=30, replace=True)
        means_data[c].append(np.mean(sample))
    
```

```python
df_means = pd.DataFrame(means_data)
df_means
```

```python
df_means.to_csv(r'raw\electricity-clt-dataset-test.csv', index=False)
```
