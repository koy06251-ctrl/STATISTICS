```python
# imports and packages
import pandas as pd
import numpy as np
```

```python
# loading dataset
df = pd.read_csv(r'C:\Users\ASUS\OneDrive\Desktop\dataset\water_quality1.csv')
```

```python
df.head()
```

```python
# storage of means
means_data = {col: [] for col in df.columns}
```

```python
np.random.seed(3000)

# sampling means
for _ in range(3000): # 3000 times
    for c in df.columns:
        sample = np.random.choice(df[c], size=30, replace=True)
        means_data[c].append(np.mean(pd.to_numeric(sample, errors='coerce')))
    
```

```python
df_means = pd.DataFrame(means_data)
df_means
```

```python
import os

os.makedirs('raw', exist_ok=True)
df_means.to_csv(r'C:\Users\ASUS\OneDrive\Desktop\dataset\water_pollution_disease_k.csv', index=False)
```
